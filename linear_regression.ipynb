```python
#importing linear regression command from "sklearn.linear_model"
!pip install scikit-learn
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import numpy as np

#imported data of ambient temperature vs soup sales
temperature = np.array(range(60, 100, 2))
temperature = temperature.reshape(-1, 1)
sales = [65, 58, 46, 45, 44, 42, 40, 40, 36, 38, 38, 28, 30, 22, 27, 25, 25, 20, 15, 5]

#now creating a linear regression called "line_fitter"
line_fitter = LinearRegression()
line_fitter = line_fitter.fit(temperature, sales)
sales_predict = line_fitter.predict(temperature)
plt.plot(temperature, sales_predict)

#plotting and showing the data + LOBF
plt.plot(temperature, sales, 'o')
plt.show()
```
